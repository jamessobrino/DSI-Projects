```python
def convert_min(x):
    if pd.isna(x):
        return 0
    x = str(x).split(':')
    if len(x) < 2:
        return int(x[0])
    else: 
        return int(x[0])*60+int(x[1])
```

```python
test_df['SECONDS']= test_df['MIN'].apply(convert_min)
```
