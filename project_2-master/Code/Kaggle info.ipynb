```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#all changed to train need to also be done on test
```

```python
train = pd.read_csv("./datasets/train.csv")
test = pd.read_csv("./datasets/test.csv")
```

```python
train.shape
```

```python
#what column is missing
set(train)- set(test)
```

```python
train.head()
```

```python
#set id as index
train.set_index("Id", inplace = True)
test.set_index("Id", inplace = True)
```

```python
#Check for null values
train.isnull().sum().sort_values(ascending = False)
```

```python
## dont do this, just subbing zeros in for na for sake of exercise fill missing values with something useful
train.fillna(0, inplace = True)
test.fillna(0, inplace = True)
```

```python
train.info()
```

Feature engineering

```python
#one - hotencode values of neighborhood
train_dummy = pd.get_dummies(train[["Neighborhood"]])
test_dummy = pd.get_dummies(test[["Neighborhood"]])
```

```python
train_dummy
```

```python
test_dummy
```

```python
# 2 neighborhoods arent in our test set
set(train_dummy) - set(test_dummy)
```

```python
missing_col_train = list(set(test_dummy) - set(train_dummy))
missing_col_test = list(set(train_dummy) - set(test_dummy))

print(f"Columns missing from train: {missing_col_train}")
print(f"Columns missing from test: {missing_col_test}")
```

```python
#filling missing neighborhood columns as zero so the train and test sets have the same size
for col in missing_col_test:
    test_dummy[col] = 0
```

```python
#check to see if it works
set(train_dummy) - set(test_dummy)
```

```python
#drop one neighborhood col to be a reference column
train_dummy.drop(columns = ['Blmngtn'], inplace = True)
test_dummy.drop(columns = ['Blmngtn'], inplace = True)
```

```python
#join the dummies on the original dataframe
train_full = train.join(train_dummy)
test_full = test.join(test_dummy)

#put all the columns in test the same order as train
test_full = test_full[train_full.drop(columns = "SalePrice").columns]
```

```python
test_full.columns
```

```python
train_full.columns
```

```python
#preprocessing and modeling
```

```python
#define features
features = ['Overall Qual', 'Lot Area']
features.extend(train_dummy.columns)
```

```python
#define X and y
X = train_full[features]
y = train_full["SalePrice"]
```

```python
#create test and train variables
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X_train.shape, X_test.shape)
```

```python
#establish baseline model for comparison
#this is a naive model
y_train_baseline = [y_train.mean()]*len(y_train)
y_test_baseline = [y_train.mean()]*len(y_test)
```

```python
print(f"baseline RMSE - train : {mean_squared_error(y_train, y_train_baseline)*.5}")
print(f"baseline RMSE - train : {mean_squared_error(y_test, y_test_baseline)*.5}")
```

```python
#LINEAR REGRESSION
lr= LinearRegression()
lr.fit(X_train, y_train)
```

```python
print(f"baseline RMSE - train : {mean_squared_error(y_train, lr.predict(X_train))*.5}")
print(f"baseline RMSE - train : {mean_squared_error(y_test, lr.predict(X_test))*.5}")
```

```python
lr.coef_
```

```python
pd.DataFrame({
    'column' : X.columns,
    'coef' : lr.coef_
}).head()
```

```python
##################################
# USE MODEL TO MAKE PREDITIONS FOR KAGGLE
X_kaggle = test_full.loc[:,features]
X_kaggle.head()
```

```python
X_kaggle["SalePrice"] = lr.predict(X_kaggle)
X_kaggle.head()
```

```python
# Set output to be dataframe with only SalePrice
output = X_kaggle[["SalePrice"]]
output.head()
```
